# Finding schools with the best math scores

```python
sorted_average_math = schools.sort_values("average_math", ascending=False)
sorted_average_math
```

```python
sorted_average_math['average_math'].max()
```

```python
max_value = (800 * 80) / 100
max_value
```

```python
best_math_schools = sorted_average_math[sorted_average_math['average_math'] >= max_value]
best_math_schools = best_math_schools[['school_name', 'average_math']]
best_math_schools
```

# Locating the NYC borough with the largest standard deviation in SAT performance

```python
schools["total_SAT"].std()
```

```python
schools["total_SAT"].mean()
```

```python
schools.groupby("borough")['school_name'].count()
```

```python
# Group by 'borough' and calculate mean and std of 'total_SAT'
borough_stats = schools.groupby('borough')['total_SAT'].agg(['mean', 'std', 'count'])
borough_stats

# # Find the borough with the largest standard deviation
largest_std_dev_borough = borough_stats['std'].idxmax()
largest_std_dev_borough

# # Extract statistics for that borough
borough_info = borough_stats.loc[largest_std_dev_borough]
print(borough_info)

# # Create the resulting DataFrame
largest_std_dev = pd.DataFrame({
    'borough': [largest_std_dev_borough],
    'num_schools': [borough_info['count']],
    'average_SAT': [round(borough_info['mean'], 2)],
    'std_SAT': [round(borough_info['std'], 2)]
})

# # Show the resulting DataFrame
print(largest_std_dev)
```
